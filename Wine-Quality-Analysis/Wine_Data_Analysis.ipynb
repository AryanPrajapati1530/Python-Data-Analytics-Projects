## Univariate Analysis

#### Quality Distribution (Bar Chart)

```python
quality_counts = df['quality'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(
    quality_counts.index,
    quality_counts,
    color='darkred'
)

plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')

plt.show()
```

#### Swarm Plot

```python
plt.figure(figsize=(10,8))

sns.swarmplot(
    x="quality",
    y="alcohol",
    data=df,
    palette='viridis'
)

plt.title('Swarm Plot for Quality and Alcohol')

plt.show()
```

###### Relationship between alcohol content and wine quality.
###### Outliers become visible.
###### Density of points shows concentration.

## Bivariate Analysis

```python
df['quality'] = df['quality'].astype(str)
```

#### Violin Plot

```python
plt.figure(figsize=(10,8))

sns.violinplot(
    x="quality",
    y="alcohol",
    data=df
)

plt.title('Violin Plot for Quality and Alcohol')

plt.show()
```

###### Distribution of alcohol for each quality level.
###### Wider sections = more observations.
###### Narrow sections = fewer observations

#### Box Plot 

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x='quality',
    y='alcohol',
    data=df
)

plt.title('Box Plot of Quality vs Alcohol')

plt.show()
```

###### Median alcohol level.
###### Spread of alcohol values.
###### Outliers.

## Multivariate Analysis

#### Correlation Heatmap

```python
plt.figure(figsize=(15,10))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    fmt='.2f',
    cmap='Pastel2',
    linewidths=2
)

plt.title('Correlation Heatmap')

plt.show()
```

###### Correlation with quality
###### Strong positive correlations (close to +1)
###### Strong negative correlations (close to -1)
###### Alcohol → positive relationship with quality
###### Volatile acidity → negative relationship with quality

#### Remove Duplicates

```python
df = df.drop_duplicates()
```

#### Verify Duplicate Removal

```python
duplicated = df.duplicated()

print(
    "Number of duplicated instances:",
    duplicated.sum()
)
```

#### Convert Quality into Binary Classification

```python
df["quality"] = np.where(
    df["quality"].astype(int) >= 7,
    1,
    0
)
```

##### Verify Classification

```python
df['quality'].value_counts()
```

###### 1 = Good Wine
###### 0 = Not Good Wine

##### Features and Label

```python
features = df.columns[:-1].values

label = [df.columns[-1]]

print("Features:", features)

print("Label:", label)
```

#### Quality Count Plot

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='quality',
    palette='BuPu'
)

plt.title('Distribution of Wine Quality')

plt.show()
```

#### Alcohol vs Quality

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x="quality",
    y="alcohol",
    data=df
)

plt.title("Alcohol vs Wine Quality")

plt.show()
```

#### Density vs Alcohol and Residual Sugar

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    x="alcohol",
    y="density",
    data=df,
    label='Alcohol'
)

sns.lineplot(
    x="residual sugar",
    y="density",
    data=df,
    label='Residual Sugar'
)

plt.title(
    "Density vs Alcohol and Residual Sugar"
)

plt.show()
```

## Insights from Wine Quality Dataset

##### Insight 1: Which Feature Has the Highest Correlation with Quality?

```python
corr_with_quality = df.corr(numeric_only=True)['quality'].sort_values(ascending=False)

print(corr_with_quality)
```

###### Find the strongest positive and negative factors affecting wine quality.

##### Insight 2: Top Positive Factors Affecting Quality

```python
corr_with_quality.sort_values(ascending=False).head(5)
```

###### Alcohol usually shows one of the strongest positive relationships.

##### Insight 3: Top Negative Factors Affecting Quality

```python
corr_with_quality.sort_values().head(5)
```

###### Volatile acidity generally has a negative effect.

##### Insight 4: Average Alcohol Content by Quality

```python
df.groupby('quality')['alcohol'].mean()
```

###### Do higher-quality wines contain more alcohol?

##### Insight 5: Average Sulphates by Quality

```python
df.groupby('quality')['sulphates'].mean()
```

###### Higher sulphate levels may be associated with better-quality wine.

##### Insight 6: Alcohol Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(df['alcohol'], bins=20, kde=True)

plt.title("Alcohol Distribution")
plt.show()
```

###### Understand the most common alcohol range.

##### Insight 7: Most Influential Features
